```python
%matplotlib inline

import numpy as np
from numpy.random import randint, choice
import cv2
from matplotlib import pyplot as plt


image = np.ones((500, 500, 3), dtype=np.uint8) * 255

for _ in range(randint(10, 15)):
    center_x = randint(0, 500)
    center_y = randint(0, 500)
    width = randint(30, 60)
    height = randint(100, 150)
    angle = choice([0,90,180,270])
    start_angle = 0
    end_angle = 180
    color = tuple(randint(0, 256, 3).tolist())

    cv2.ellipse(image, (center_x, center_y), (width, height), angle, start_angle, end_angle, color, -1)

border_size = 2
border_image = np.pad(image, ((border_size, border_size), (border_size, border_size), (0, 0)), mode='constant', constant_values=0)

fig = plt.imshow(border_image)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
        
```
