```python
%matplotlib inline

import numpy as np
from numpy.random import randint, choice
import cv2
from matplotlib import pyplot as plt

image = np.ones((2048, 2048, 3), dtype=np.uint8) * 255

for _ in range(randint(5, 10)):
    center_x = randint(0, 2048)
    center_y = randint(0, 2048)
    width = randint(200, 250)
    height = randint(700, 1300)
    angle = choice([0,90,180,270])
    start_angle = 0
    end_angle = 180

    color = tuple(randint(0, 256, 3).tolist())

    cv2.ellipse(image, (center_x, center_y), (width, height), angle, start_angle, end_angle, color, -1)

border_size = 1
border_image = np.pad(image, ((border_size, border_size), (border_size, border_size), (0, 0)), mode='constant', constant_values=1)

fig = plt.imshow(border_image)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
        
```

```python
cv2.imwrite("06.jpg", border_image)
```
